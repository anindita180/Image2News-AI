```python
import os
import cv2
import easyocr
import torch
import json

# Define dataset path
image_folder = "/content/drive/MyDrive/Delta_Headlines"
ocr_output_file = "ocr_results.json"

# Create OCR Reader
reader = easyocr.Reader(["en"], gpu=torch.cuda.is_available())

# Function to extract text
def extract_text_from_images():
    results = {}
    for img_file in os.listdir(image_folder):
        if img_file.endswith(('.png', '.jpg', '.jpeg')):
            img_path = os.path.join(image_folder, img_file)
            text = reader.readtext(img_path, detail=0)
            results[img_file] = " ".join(text)  # Save text from each image
            print(f"Extracted text from {img_file}")

    # Save results for next step
    with open(ocr_output_file, "w") as f:
        json.dump(results, f)

    print(f"OCR completed! Results saved in {ocr_output_file}")

# Run OCR
extract_text_from_images()
```

```python
import json
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load OCR results
ocr_output_file = "ocr_results.json"
summarized_output_file = "summarized_results.json"

# Load Pretrained Summarization Model
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Load extracted text
with open(ocr_output_file, "r") as f:
    extracted_texts = json.load(f)

summarized_results = {}

# Summarize each extracted text
for img_file, text in extracted_texts.items():
    inputs = tokenizer("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(inputs.input_ids, max_length=150, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    summarized_results[img_file] = summary
    print(f"Summarized {img_file}")

# Save results for next step
with open(summarized_output_file, "w") as f:
    json.dump(summarized_results, f)

print(f"Summarization completed! Results saved in {summarized_output_file}")
```

```python
import json
from transformers import pipeline

# Load summarized text
summarized_output_file = "summarized_results.json"
news_output_file = "news_results.json"

# Load text generation model
news_generator = pipeline("text-generation", model="EleutherAI/gpt-neo-1.3B", trust_remote_code=True)

# Load summaries
with open(summarized_output_file, "r") as f:
    summaries = json.load(f)

news_results = {}

# Generate news reports
for img_file, summary_text in summaries.items():
    generated_text = news_generator(summary_text, max_length=200, num_return_sequences=1)[0]['generated_text']
    news_results[img_file] = generated_text
    print(f"Generated news for {img_file}")

# Save results for next step
with open(news_output_file, "w") as f:
    json.dump(news_results, f)

print(f"News generation completed! Results saved in {news_output_file}")
```
